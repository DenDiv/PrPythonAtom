```python
class HashMap:
    class Entry:
        def __init__(self, key, value):
            # TODO store key and value
            self.key = key
            self.value = value
            self.next = None  # указатель на след. элем.

        def get_key(self):
            # TODO return key
            return self.key

        def get_value(self):
            # TODO return value
            return self.value

        def __eq__(self, other):
            # TODO реализовать функцию сравнения
            return self.key == other.key

    def __init__(self, bucket_num=64):
        # TODO Создать массив для бакетов
        self.arr = [None for i in range(bucket_num)]

    def get(self, key, default_value=None):
        # TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        hash = self._get_hash(key)
        index = self._get_index(hash)
        our_elem = self.arr[index]
        while our_elem is not None:
            if our_elem.get_key() == key:
                return our_elem.get_value()
            else:
                our_elem = our_elem.next
        return default_value

    def put(self, key, value):
        # TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        hash = self._get_hash(key)
        index = self._get_index(hash)
        new_elem = self.Entry(key, value)
        if self.arr[index] is None:
            self.arr[index] = new_elem
        else:
            our_elem = self.arr[index]
            while our_elem is not None:
                # если ключи совпадают
                if our_elem == new_elem:
                    our_elem.value = new_elem.value
                    break
                else:
                    our_elem = our_elem.next
            # добавляем Entry в начало списка, если в нем нет эл-та с таким же ключем
            if our_elem is None:
                new_elem.next = self.arr[index]
                self.arr[index] = new_elem

    def __len__(self):
        # TODO Возвращает количество Entry в массиве
        full_counter = 0
        for i in self.arr:
            our_elem = i
            counter = 0
            while our_elem is not None:
                counter += 1
                our_elem = our_elem.next
            full_counter += counter
        return full_counter

    def _get_hash(self, key):
        # TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key) % len(self.arr)

    def _get_index(self, hash_value):
        # TODO По значению хеша вернуть индекс элемента в массиве
        return hash_value

    def values(self):
        pass

    def keys(self):
        pass

    def items(self):
        pass

    def __str__(self):
        # формат вывода: { bucket_0: key_0, val_0; key_1, val_1 ... ; bucket_1: key_0, val_0 ... }
        str_ = '{'
        for i in range(len(self.arr)):
            str_ += str(i) + ': '
            our_elem = self.arr[i]
            if our_elem is None:
                str_ += 'None' + '; '
            else:
                while our_elem is not None:
                    str_ += str(our_elem.get_key()) + ', ' + str(our_elem.get_value()) + ';'
                    our_elem = our_elem.next
        str_ += '}'
        return str_
```
